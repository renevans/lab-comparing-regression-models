```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import re
```

```python
# Load cleaned file from previous lab (lab-data-cleaning-and-wrangling)

data= pd.read_csv('customer_df_cleaned.csv')
```

```python
# display(data)
data.head()
# data.info()
```

```python
data = data.drop(['Unnamed: 0'], axis = 1)
```

```python
data.T
```

```python
# 1.  In this final lab, we will model our data. Import sklearn train_test_split and separate the data.

from sklearn.model_selection import train_test_split
```

```python
X = data.drop(['total_claim_amount'],axis=1)
y = data['total_claim_amount']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=100)
```

```python
# 2.  Try a simple linear regression with all the data to see whether we are getting good results.
from sklearn import linear_model
LR = linear_model.LinearRegression()
LR.fit(X_train,y_train)
```

```python
score = LR.score(X_test, y_test)
score
```

```python
# 3.  Great! Now define a function that takes a list of models and train (and tests) them so we can try a lot of them without repeating code.
```

```python
# Importing and giving an alias to the models
from sklearn.neighbors import KNeighborsRegressor
KNN = KNeighborsRegressor(n_neighbors=5)
```

```python
# Function that takes a list of models
scores = []
models = [LR, KNN] # insert model acronym here
for x in models:
  x.fit(X_train,y_train)
  scores.append(x.score(X_test, y_test))  
```

```python
# 4.  Use the function to check LinearRegressor and KNeighborsRegressor.
```

```python
d = {'model':models,'score':scores}
results = pd.DataFrame(d)
results
```

```python
# 5.  You can check also the MLPRegressor for this task!

from sklearn.neural_network import MLPRegressor
MLP = MLPRegressor()
scores = []
models = [LR, KNN, MLP] # insert model acronym here
for x in models:
  x.fit(X_train,y_train)
  scores.append(x.score(X_test, y_test))     
d = {'model':models,'score':scores}
results = pd.DataFrame(d)
```
